```python
import PyPDF2
import icd10
import re
from treelib import Tree,Node
```

## Read in the ICD10 file

```python
icd=open('icd10cm_index_2022.pdf',mode="rb")
icd=PyPDF2.PdfFileReader(icd)
pages=icd.numPages #get the number of pages
```

## Save as txt file

```python
lst_without=[]
for i in range(pages):
    PageObj=icd.getPage(i)
    #lst.append({'Page Number :':i})
    #print('Page Number :',i)
    text=PageObj.extractText().split('   ')
    for j in range(len(text)):
        #lst.append(text[j])
        for k in text[j].split('\n'):
            if len(k)!=0:
                k=k.replace('-','') 
                lst_without.append(k)
```

```python
with open("./output.txt", "w") as txt_file:
     for line in lst_without:
          txt_file.write(line + "\n")  # works with any number of elements in a line
```
